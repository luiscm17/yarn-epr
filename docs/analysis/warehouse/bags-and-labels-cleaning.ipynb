# Bags and labels — Data Cleaning & Normalization

Analyze and normalize the `database-bags-and-labels-sic-jac.csv` reference data from SIC JAC.
Focus on columns A (codigo), B (descripcion), and L (SUB-CATEGORIA).

**Source:** `docs/research/data/database-bags-and-labels-sic-jac.csv`

# Load and Overview 

```python
# Leer el archivo completo como texto
import os
print(f"Working directory: {os.getcwd()}")

# Usar ruta absoluta
abs_path = r"c:\Users\Alexander\Desktop\illampu\yarn-epr\docs\research\data\database-bags-and-labels-sic-jac.csv"
with open(abs_path, 'r', encoding='utf-8') as f:
    contenido = f.read()
print(contenido)
```

```python
# Extraer los items y separarlos por prefijo de código
items_section = contenido.split("ITEMS:")[1]

# Diccionario para agrupar items por prefijo
items_by_prefix = {}

# Procesar cada línea
for line in items_section.strip().split('\n'):
    line = line.strip()
    if line:
        # Extraer el prefijo (BLS-, ETQ-, DEV-, TAL-, etc.)
        parts = line.split()
        if parts:
            codigo = parts[0]
            prefijo = codigo.split('-')[0]
            
            if prefijo not in items_by_prefix:
                items_by_prefix[prefijo] = []
            
            items_by_prefix[prefijo].append(line)

# Mostrar items organizados por prefijo
for prefijo in sorted(items_by_prefix.keys()):
    print(f"\n{'='*60}")
    print(f"PREFIJO: {prefijo}")
    print(f"{'='*60}")
    for item in items_by_prefix[prefijo]:
        print(item)
```

## Análisis Detallado: Bolsas y Etiquetas

Separar por tipo de producto y analizar patrones en códigos y descripciones.

```python
# Crear DataFrames para cada prefijo
import pandas as pd

# Función para parsear un item
def parse_item(line):
    parts = line.split(None, 1)  # Dividir en máximo 2 partes
    if len(parts) == 2:
        codigo, descripcion = parts
        return {'codigo': codigo, 'descripcion': descripcion}
    return None

# Crear DataFrames separados
dfs = {}
for prefijo in sorted(items_by_prefix.keys()):
    parsed_items = [parse_item(item) for item in items_by_prefix[prefijo]]
    parsed_items = [item for item in parsed_items if item is not None]
    dfs[prefijo] = pd.DataFrame(parsed_items)

# Mostrar resumen
print("=" * 70)
print("RESUMEN DE PRODUCTOS")
print("=" * 70)
for prefijo, df_tmp in dfs.items():
    print(f"\n{prefijo}: {len(df_tmp)} items")
    print(df_tmp.to_string())
```

```python
## Análisis detallado de Bolsas (BLS)

bolsas_df = dfs['BLS'].copy()

print("=" * 70)
print("ANÁLISIS: BOLSAS")
print("=" * 70)

# Extraer información de las descripciones de bolsas
# Patrón: BOLSA COLOR DIMENSIONES o BOLSA TIPO DIMENSIONES

def parse_bolsa(desc):
    # Intentar extraer dimensiones (números con *)
    import re
    dim_match = re.search(r'(\d+)\*(\d+)\*(\d+)', desc)
    
    # Extraer el tipo/color (primera palabra después de BOLSA)
    type_match = re.search(r'BOLSA\s+([A-Z\s]+?)(?:\d|\(|$)', desc)
    
    return {
        'type': type_match.group(1).strip() if type_match else '',
        'dimensions': dim_match.group(0) if dim_match else '',
        'description': desc
    }

bolsas_parsed = bolsas_df['descripcion'].apply(parse_bolsa)
bolsas_df = pd.concat([bolsas_df, bolsas_parsed.apply(pd.Series)], axis=1)

print("\nBolsas con información extraída:")
print(bolsas_df[['codigo', 'type', 'dimensions']].to_string())

print("\n\nDimensiones únicas en bolsas:")
print(bolsas_df['dimensions'].value_counts())

print("\n\nTipos/Colores únicos:")
print(bolsas_df['type'].unique())
```

```python
## Análisis detallado de Etiquetas (ETQ)

etiquetas_df = dfs['ETQ'].copy()

print("=" * 70)
print("ANÁLISIS: ETIQUETAS")
print("=" * 70)

# Extraer información de las descripciones de etiquetas
# Patrón: MARCA COLOR/TIPO o MARCA SERIE

def parse_etiqueta(desc):
    import re
    
    # Identificar marca (COLIBRI, KOLLITA, AMIGURUMI)
    marca = ''
    tipo = ''
    
    if 'COLIBRI' in desc:
        marca = 'COLIBRI'
        tipo = desc.replace('COLIBRI', '').replace('ETIQUETA', '').strip()
    elif 'KOLLITA' in desc:
        marca = 'KOLLITA'
        tipo = desc.replace('KOLLITA', '').strip()
    elif 'AMIGURUMI' in desc:
        marca = 'AMIGURUMI'
        tipo = desc.replace('ETIQUETA AMIGURUMI', '').strip()
    else:
        tipo = desc
    
    return {
        'marca': marca,
        'tipo_detalle': tipo if tipo else '(sin especificar)',
        'description': desc
    }

etiquetas_parsed = etiquetas_df['descripcion'].apply(parse_etiqueta)
etiquetas_df = pd.concat([etiquetas_df, etiquetas_parsed.apply(pd.Series)], axis=1)

print("\nEtiquetas con información extraída:")
print(etiquetas_df[['codigo', 'marca', 'tipo_detalle']].to_string())

print("\n\nMarcas encontradas:")
print(etiquetas_df['marca'].value_counts())

print("\n\nTipos/Variantes por marca:")
for marca in etiquetas_df['marca'].unique():
    if marca:
        print(f"\n{marca}:")
        tipos = etiquetas_df[etiquetas_df['marca'] == marca]['tipo_detalle'].unique()
        for tipo in tipos:
            print(f"  - {tipo}")
```
